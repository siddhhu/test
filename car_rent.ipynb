```python
Y=df['Monthly Payment']
```

```python
df2 = df
drops = ['Monthly Payment']
df2 = df2.drop(drops,axis=1)
df2.dtypes
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(df2, Y,
                                                      train_size=0.7, test_size=0.3,
                                                      random_state=0)
```

```python
# import joblib
# # Save the model as a pickle in a file 
# joblib.dump(clf, 'car_new1.pkl') 
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error


clf1=RandomForestRegressor(n_estimators=450)

clf1.fit(X_train,y_train)


pred = clf1.predict(X_valid)
```

```python
X_train
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(pred, y_valid)
```

```python
from sklearn.metrics import r2_score
r2_score(pred, y_valid)
```
